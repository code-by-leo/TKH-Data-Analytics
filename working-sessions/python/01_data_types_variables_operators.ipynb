## Why this matters for data science

When pandas reads a CSV file, it guesses a dtype for every column: `int64`, `float64`, `object`, or `bool`. Getting that guess wrong is one of the most common sources of bugs in data pipelines. A salary column stored as `object` instead of `float64` will silently break every aggregation you run on it. Knowing Python's type system is the first step toward diagnosing and fixing those issues.

## What's happening?

In Python, every value has a **type** that determines what you can do with it. You do not declare types explicitly; Python infers them from the value you assign.

| Type | Example | Typical use in data science |
|------|---------|----------------------------|
| `int` | `42` | Counts, indices, discrete labels |
| `float` | `3.14` | Measurements, probabilities, prices |
| `str` | `"Alice"` | Names, categories, raw text |
| `bool` | `True` | Flags, filter masks, binary targets |
| `NoneType` | `None` | Missing values before loading into pandas |

```python

salary = 72_500.0    # float
name   = "Alice"     # str

```

### Operators behave differently depending on type

The `+` operator adds two numbers but *concatenates* two strings. Mixing types without an explicit conversion raises a `TypeError`. This is intentional: Python refuses to guess what `"3" + 2` should mean.

Return to the widget and try selecting the string `"42"` versus the integer `42`, then apply the `*` operator to both. The results will surprise you.

## Real-world example: Column dtypes after loading a CSV

When you load a CSV with `pd.read_csv()`, pandas assigns a dtype to every column. Getting these right before analysis is a routine first step in every data science project.

Notice:

- **Notice:** String columns like `name` and `department` become `object` dtype in pandas, which is just Python `str` under the hood
- **Notice:** The `active` flag is read as `bool` only if the CSV contains `True`/`False` literally; `"yes"`/`"no"` would become `object` instead
- **Notice:** `start_date` comes in as `object` by default; you must call `pd.to_datetime()` explicitly, which is why type awareness matters from the very first line of a pipeline

> **Discussion question:** A colleague loads a CSV and finds that a `salary` column has dtype `object`. Before writing any code, what are two plausible reasons this happened, and how would you check each one?

````python
# ── Dtype audit for a realistic employee dataset ────────────────────────────────
data = {
    "employee_id": [101, 102, 103, 104, 105],
    "name":        ["Alice", "Bob", "Carol", "David", "Eva"],
    "department":  ["Engineering", "Marketing", "Engineering", "HR", "Marketing"],
    "salary":      [95000.0, 72000.0, 88000.0, 61000.0, 78500.0],
    "tenure_yrs":  [3, 7, 2, 10, 5],
    "active":      [True, True, False, True, True],
    "start_date":  ["2021-03-15", "2017-06-01", "2022-09-10", "2014-01-20", "2019-11-05"],
}
df = pd.DataFrame(data)

print(df.dtypes)
````

**Errata**

Historically, Pandas used NumPy as its primary backend. Because NumPy doesn't have a dedicated text/string data type, it lumps all text (and mixed data) into a generic Python object pointers array.

However, in newer versions of Pandas, you will often see text columns explicitly labeled as str. Modern Python environments use an upgraded engine under the hood that handles text much faster and more efficiently, allowing Pandas to label strings as `str` instead of `object`

### Common dtype mismatches and how to fix them

| Symptom | Likely cause | Fix |
|---------|-------------|-----|
| Numeric column reads as `object` | Non-numeric characters in the column (e.g., `"$72,000"`) | Strip characters, then `pd.to_numeric()` |
| Date column reads as `object` | pandas does not auto-parse dates by default | `pd.to_datetime(df["col"])` |
| Boolean column reads as `object` | Values are `"yes"`/`"no"` or `"1"`/`"0"` strings | Map with `.map({"yes": True, "no": False})` |
| Integer column reads as `float64` | Column has at least one missing value (`NaN`) | Expected; fill then cast with `.astype(int)` |
